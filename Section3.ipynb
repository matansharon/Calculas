```python
#solve this equation with numpy: 3-5x+(5/6)*x^2+12.4*x^3=0
#first, we need to define the function
def f(x):
    return 3-5*x+(5/6)*x**2+12.4*x**3

#then, we need to define the derivative of the function
def df(x):
    return -5+(5/3)*x+37.2*x**2

#now, we need to define the newton method
def newton(x0, tol, maxiter):
    x = x0
    for i in range(maxiter):
        x1 = x - f(x)/df(x)
        if abs(x1-x) < tol:
            return x1
        x = x1
    return x

#now, we need to define the secant method
def secant(x0, x1, tol, maxiter):
    for i in range(maxiter):
        x2 = x1 - f(x1)*(x1-x0)/(f(x1)-f(x0))
        if abs(x2-x1) < tol:
            return x2
        x0 = x1
        x1 = x2
    return x2

#now, we need to define the bisection method
def bisection(a, b, tol, maxiter):
    for i in range(maxiter):
        c = (a+b)/2
        if abs(f(c)) < tol:
            return c
        if f(a)*f(c) < 0:
            b = c
        else:
            a = c
    return c
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a curve that does not represent a function
x1 = np.linspace(-1, 1, 100)
y1 = np.sqrt(1 - x1 ** 2)
y2 = -np.sqrt(1 - x1 ** 2)

# Plot the curve
plt.plot(x1, y1, "b")
plt.plot(x1, y2, "b")
plt.axvline(x=0, color="r")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Vertical line test: Not a function")
plt.show()

# Define a curve that represents a function
x2 = np.linspace(-5, 5, 100)
y3 = x2 ** 2

# Plot the curve
plt.plot(x2, y3, "b")
plt.axvline(x=0, color="r")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Vertical line test: A function")
plt.show()
```

```python


# coefficients of the polynomial
coeff = [12.4, 5/6, -5, 3]
# compute roots
roots = np.roots(coeff)

# generate x values
x = np.linspace(-2, 2, 400)

# calculate y values
y = coeff[0]*x**3 + coeff[1]*x**2 + coeff[2]*x + coeff[3]

plt.figure(figsize=(10,6))
# plot the function
plt.plot(x, y, label='12.4*x^3 + (5/6)*x^2 -5x + 3 = 0')
# plot the roots
plt.scatter(np.real(roots), np.zeros(len(roots)), color='red') # we use np.real to ignore imaginary part of roots
plt.title('Plot of the Polynomial and its Roots')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```

```python


# Define the function
def f(z):
    return np.exp(z**2)

# Create a grid of points
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x, y)
Z = X + 1j*Y

# Evaluate the function at each point
F = f(Z)

# Create the magnitude plot
plt.figure()
plt.imshow(np.abs(F), extent=(x.min(), x.max(), y.min(), y.max()), origin='lower')
plt.colorbar()
plt.title('Magnitude of f(z)')

# Create the phase plot
plt.figure()
plt.imshow(np.angle(F), extent=(x.min(), x.max(), y.min(), y.max()), origin='lower')
plt.colorbar()
plt.title('Phase of f(z)')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return np.where(x <= 0, 2*x + 1, x**2)

# Create an array of x values
x = np.linspace(-10, 10, 400)

# Evaluate the function at each x
y = f(x)

# Create the plot
plt.figure()
plt.legend()
plt.plot(x, y)
plt.title('Plot of the piecewise function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

```python


# Define the function
def f(x):
    return x**2 + 3*x**3 - x**4

# Create an array of x values
x = np.linspace(-2, 2, 400)

# Evaluate the function at each x
y = f(x)

# Create the plot
plt.figure()
plt.plot(x, y, label='$y = x^2 + 3x^3 - x^4$')  # Use LaTeX syntax for the label
plt.title('Plot of the function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()  # Add the legend
plt.grid(True)
plt.show()
```

```python


# Define the function
def f(beta):
    return -beta**4 + 3*beta**3 + beta**2

# Create an array of beta values
beta = np.linspace(-2, 2, 400)

# Evaluate the function at each beta
y = f(beta)

# Create the plot
#mark the lines of x=0 and y=0 with red color

plt.figure()
plt.plot(beta, y, label='$f(\\beta) = -\\beta^4 + 3\\beta^3 + \\beta^2$')
plt.title('Plot of the function')
plt.xlabel('$\\beta$')
plt.ylabel('f($\\beta$)')
plt.legend()  # Add the legend
plt.grid(True)
plt.axhline(y=0, color='r', linestyle='-')
plt.axvline(x=0, color='r', linestyle='-')
plt.show()
```
