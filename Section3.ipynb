```python
#solve this equation with numpy: 3-5x+(5/6)*x^2+12.4*x^3=0
#first, we need to define the function
def f(x):
    return 3-5*x+(5/6)*x**2+12.4*x**3

#then, we need to define the derivative of the function
def df(x):
    return -5+(5/3)*x+37.2*x**2

#now, we need to define the newton method
def newton(x0, tol, maxiter):
    x = x0
    for i in range(maxiter):
        x1 = x - f(x)/df(x)
        if abs(x1-x) < tol:
            return x1
        x = x1
    return x

#now, we need to define the secant method
def secant(x0, x1, tol, maxiter):
    for i in range(maxiter):
        x2 = x1 - f(x1)*(x1-x0)/(f(x1)-f(x0))
        if abs(x2-x1) < tol:
            return x2
        x0 = x1
        x1 = x2
    return x2

#now, we need to define the bisection method
def bisection(a, b, tol, maxiter):
    for i in range(maxiter):
        c = (a+b)/2
        if abs(f(c)) < tol:
            return c
        if f(a)*f(c) < 0:
            b = c
        else:
            a = c
    return c
```

```python


# Define the function
def f(z):
    return np.exp(z**2)

# Create a grid of points
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x, y)
Z = X + 1j*Y

# Evaluate the function at each point
F = f(Z)

# Create the magnitude plot
plt.figure()
plt.imshow(np.abs(F), extent=(x.min(), x.max(), y.min(), y.max()), origin='lower')
plt.colorbar()
plt.title('Magnitude of f(z)')

# Create the phase plot
plt.figure()
plt.imshow(np.angle(F), extent=(x.min(), x.max(), y.min(), y.max()), origin='lower')
plt.colorbar()
plt.title('Phase of f(z)')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Define the symbolic function
x = sp.symbols('x')
fun_sym = x + x**2 + x**3

# Convert the symbolic function to a numerical function
fun = sp.lambdify(x, fun_sym, 'numpy')

# Create an array of x values
x_values = np.linspace(-2, 2, 400)

# Evaluate the function at each x
y_values = fun(x_values)

# Create the plot
plt.figure()
plt.plot(x_values, y_values)
plt.title('Plot of the function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python


# Define the symbolic variables
x, y = sp.symbols('x y')

# Define the symbolic polynomials
poly1 = x**5 + 2*x*y + y**2
poly2 = x - 3*x*y

# Multiply the polynomials
result = sp.expand(poly1 * poly2)

# Convert the symbolic function to a numerical function
f = sp.lambdify((x, y), result, 'numpy')

# Create a grid of points
x_values = np.linspace(-2, 2, 100)
y_values = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x_values, y_values)

# Evaluate the function at each point
Z = f(X, Y)

# Create the plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title('Plot of the function')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Define the function to compute the Maclaurin series
def maclaurin_e(x, order):
    result = 0
    for n in range(order + 1):
        if n == 0:
            result += x**n
        else:
            result += -1**(n+1) *(x**(2*n))/ math.factorial(2*n-1)
    return np.sin(result)

# Create an array of x values
x = np.linspace(-2, 2, 400)

# Compute the Maclaurin series up to different orders
orders = [1, 2, 3, 4, 5]
for order in orders:
    y = maclaurin_e(x, order)
    plt.plot(x, y, label=f'order {order}')

# Plot the actual function for comparison
y_actual = np.exp(x)
plt.plot(x, y_actual, label='actual function', linestyle='--')

plt.title('Maclaurin series approximation of $e^x$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
